```python
with open("../data/source/en-fr/europarl/europarl-v7.fr-en.en", "r", encoding="utf-8") as f:
    europarl = f.read().split("\n")
data = []
for e in europarl:
    tokens = e.split(" ")
    data.append(len(tokens))
```

```python
from matplotlib import pyplot as plt 
import numpy as np  

# Creating plot
fig = plt.figure(figsize =(10, 7))
  
plt.hist(data, bins = range(0, 101, 10)) 
  
plt.title("EuroParl Histogram") 
  
# show plot
plt.show()
```
